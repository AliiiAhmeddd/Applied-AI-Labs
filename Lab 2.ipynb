What is NumPy, and what is its primary purpose in Python?

NumPy (Numerical Python) is a Python library used for working with numerical data.
Its primary purpose is to provide:

Powerful support for mathematical operations

Fast array and matrix manipulation

Efficient handling of large datasets

NumPy arrays are much faster and more memory-efficient than Python lists, which makes NumPy essential for data science and machine learning.

What is Pandas, and what is its primary purpose in Python?

Pandas is a Python library designed for data analysis and data manipulation.
Its primary purpose is to provide:

DataFrames, which allow you to store and work with data in table format (rows & columns)

Tools for cleaning, filtering, grouping, merging, and transforming datasets

Easy loading of data from CSV, Excel, SQL, JSON, and more

Pandas is one of the most important libraries for preparing and analysing data before it is used in machine learning models.

```python
import pandas as pd

sales_data = pd.DataFrame({
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"],
    "Revenue": [1000, 1200, None, None, 1500]
})

print(sales_data)
```

```python
sales_data["Revenue"] = sales_data["Revenue"].fillna(800)
print(sales_data)
```

Data Preparation Basics
Part 3 - Removing duplicates

```python
#Task 1

import pandas as pd

orders_df = pd.DataFrame({
    "OrderID": [101, 102, 103, 104, 105],
    "CustomerID": ["A100", "A101", "A100", "A102", "A103"],
    "Product": ["Laptop", "Smartphone", "Laptop", "Tablet", "Smartphone"]
})

print(orders_df)
```

```python
#Task 2

# Check for duplicates based on OrderID AND CustomerID
duplicates = orders_df.duplicated(subset=["OrderID", "CustomerID"])
print("Duplicate rows:")
print(orders_df[duplicates])

# Drop the duplicates
orders_df = orders_df.drop_duplicates(subset=["OrderID", "CustomerID"])

print("\nDataFrame after removing duplicates:")
print(orders_df)
```

Data Preparation Basics
Part 4 - Concatenating and transforming data

```python
import pandas as pd

# Dataset 1
sales_data_1 = pd.DataFrame({
    "OrderID": [101, 102, 103],
    "CustomerID": ["A100", "A101", "A100"],
    "Product": ["Laptop", "Smartphone", "Tablet"]
})

# Dataset 2
sales_data_2 = pd.DataFrame({
    "OrderID": [104, 105, 106],
    "CustomerID": ["A102", "A103", "A101"],
    "Product": ["Monitor", "Keyboard", "Mouse"]
})

# Concatenate vertically (row-wise)
combined_sales_data = pd.concat([sales_data_1, sales_data_2], ignore_index=True)

print(combined_sales_data)
```
