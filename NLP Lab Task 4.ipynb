Ex 1

```python
# ----------------------------------------------------
# Exercise 1: Generate Trigrams
# ----------------------------------------------------

from nltk import ngrams
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

# Sample text
text = "Natural Language Processing with Python"

# Tokenize the text into words
tokens = nltk.word_tokenize(text)

# Generate trigrams
trigrams = list(ngrams(tokens, 3))

print("Trigrams:")
for grams in trigrams:
    print(grams)
```

Challenge 1

```python
# ----------------------------------------------------
# Challenge 1: Trigram Frequency
# ----------------------------------------------------

from collections import Counter

# Count trigram occurrences
trigram_freq = Counter(trigrams)

print("\nTrigram Frequencies:")
for trigram, count in trigram_freq.items():
    print(f"{trigram}: {count}")
```

```python
# ----------------------------------------------------
# Exercise 2: Bigram Language Model
# ----------------------------------------------------

from collections import defaultdict
from nltk import ngrams
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

# Sample text corpus
corpus = [
    "Natural Language Processing is fascinating.",
    "Language models are important in NLP.",
    "Machine learning and NLP are closely related."
]

# Tokenize the text into words
tokenized_corpus = [nltk.word_tokenize(sentence) for sentence in corpus]

# Function to train bigram model
def train_bigram_model(tokenized_corpus):
    model = defaultdict(lambda: defaultdict(lambda: 0))

    # Count bigrams
    for sentence in tokenized_corpus:
        for w1, w2 in ngrams(sentence, 2):
            model[w1][w2] += 1

    # Convert counts to probabilities
    for w1 in model:
        total_count = float(sum(model[w1].values()))
        for w2 in model[w1]:
            model[w1][w2] /= total_count
    
    return model

# Train the bigram model
bigram_model = train_bigram_model(tokenized_corpus)

# Function to get probability of a bigram
def get_bigram_probability(bigram_model, w1, w2):
    return bigram_model[w1][w2]

# Probability of ("Language", "Processing")
print("\nBigram Probability (Processing | Language):")
print(get_bigram_probability(bigram_model, 'Language', 'Processing'))
```

Challenge 2

```python
# ----------------------------------------------------
# Challenge 2: Bigram Probability ('models', 'are')
# ----------------------------------------------------

prob_models_are = get_bigram_probability(bigram_model, 'models', 'are')

print("\nBigram Probability ('models', 'are'):")
print(prob_models_are)
```
