Challenge Solution 

```python
from textblob import TextBlob

# Challenge sentence
challenge_sentence = "The new update improved some features, but it introduced several annoying bugs, making the overall experience frustrating."

# Analyze using TextBlob
blob = TextBlob(challenge_sentence)
sentiment = blob.sentiment

print("Challenge Sentence:")
print(challenge_sentence)
print("\nTextBlob Sentiment Analysis:")
print(f"Polarity: {sentiment.polarity:.3f}")
print(f"Subjectivity: {sentiment.subjectivity:.3f}")

if sentiment.polarity > 0:
    label = "Positive"
elif sentiment.polarity < 0:
    label = "Negative"
else:
    label = "Neutral"

print(f"\nDetected Sentiment: {label}")
```

What sentiment did TextBlob detect, and why?

TextBlob assigns sentiment based on the polarity of individual words and phrases.
For this sentence:

"The new update improved some features, but it introduced several annoying bugs, making the overall experience frustrating."

TextBlob sees:

Positive words: "improved"

Negative words: "annoying", "bugs", "frustrating"

Since the negative words carry stronger polarity values, the overall polarity becomes negative, so TextBlob classifies it as Negative sentiment.

Limitations of TextBlob for this kind of sentence

TextBlob is a rule-based, dictionary-driven sentiment analyzer.
It struggles with:

- Mixed or contrasting sentiment
The sentence contains both praise and criticism. TextBlob does not understand structure, so it averages them.

- Context and nuance
It cannot understand that the negative part is more impactful than the positive part.

- Intensity differences
Words like "frustrating" may not be weighted strongly enough.

- Sarcasm & irony
Rule-based methods cannot detect sarcasm such as:

“Great job breaking the app again!”

Which method is best for complex reviews?
1. TextBlob

- Very easy to use

- Good for simple sentiment

- Poor with nuance, mixed opinions, or long reviews

2. AFINN

- Slightly better weighting of negative words

- Still dictionary-based, no understanding of context

- Can't handle long or complex sentences well

3. Logistic Regression (Machine Learning Model) — BEST CHOICE

- Learns patterns from data

- Handles more complex expressions

- Understands weighting based on real examples

- Performs better on reviews, nuanced language, multi-sentence inputs

- Requires labelled training data

- Performance depends on dataset size and quality

Conclusion

Logistic Regression (with TF-IDF features) is the best method among the three for analyzing long, complex reviews because:

It learns from real examples

It better captures patterns of negative vs positive sentiment

It is less sensitive to ambiguous or mixed language

It performs better on multi-clause, nuanced opinions
